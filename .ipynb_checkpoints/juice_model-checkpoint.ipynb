# Classification des Qualités de Vins

Ce notebook présente une analyse complète de classification des qualités de Vins en utilisant différents modèles de machine learning(XGBoost, Kernel SVM).

## Table des Matières
1. Configuration et Chargement des Données
2. Préparation des Données
3. Modélisation et Évaluation
4. Optimisation des Modèles
5. Analyse des Résultats
6. Conclusions

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split, RandomizedSearchCV, cross_val_score, learning_curve, StratifiedKFold
from sklearn.preprocessing import StandardScaler
from sklearn.svm import SVC
from sklearn.pipeline import Pipeline
from xgboost import XGBClassifier
from sklearn.metrics import accuracy_score, precision_score, recall_score,f1_score, confusion_matrix, classification_report


import joblib
import warnings
warnings.filterwarnings('ignore')

# Configuration
plt.style.use('seaborn-v0_8')
sns.set_palette('husl')
```

```python
data = pd.read_csv('juice_data.csv')
data.head()
```

```python
data.shape
```

```python
X = data.drop(['quality', 'quality_category'], axis=1)
y = data['quality_category']

X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.2, random_state=123, stratify=y
)

print(f"Taille train: {X_train.shape}, taille test: {X_test.shape}")
```

```python
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)

print(f"Train set: {X_train.shape}")
print(f"Test set: {X_test.shape}")

```

```python
# 2. Fonction d'évaluation
def evaluate_model(model, X_train, y_train, X_test, y_test, model_name):
    model.fit(X_train, y_train)
    y_train_pred = model.predict(X_train)
    y_test_pred = model.predict(X_test)
    
    metrics = {
        'Model': model_name,
        'Train_Accuracy': accuracy_score(y_train, y_train_pred),
        'Test_Accuracy': accuracy_score(y_test, y_test_pred),
        'Precision': precision_score(y_test, y_test_pred, average='weighted', zero_division=0),
        'Recall': recall_score(y_test, y_test_pred, average='weighted', zero_division=0),
        'F1-Score': f1_score(y_test, y_test_pred, average='weighted', zero_division=0)
    }
    return metrics
```

```python
xgb_model = XGBClassifier(random_state=123, use_label_encoder=False, eval_metric='mlogloss', n_jobs=-1)
svm_model = SVC(kernel='rbf', probability=False, random_state=123)
```

```python
# Évaluation
xgb_results = evaluate_model(xgb_model, X_train_scaled, y_train, X_test_scaled, y_test, 'XGBoost')
svm_results = evaluate_model(svm_model, X_train_scaled, y_train, X_test_scaled, y_test, 'SVM (RBF)')

results_df = pd.DataFrame([xgb_results, svm_results])
display(results_df)
```

```python
best_idx = results_df['F1-Score'].idxmax()
best_model_name = results_df.loc[best_idx, 'Model']
best_f1_score = results_df.loc[best_idx, 'F1-Score']
print(f"Meilleur modèle initial: {best_model_name} (F1-Score = {best_f1_score:.4f})")
```

```python
# Vérification overfitting
train_acc = results_df.loc[best_idx, 'Train_Accuracy']
test_acc = results_df.loc[best_idx, 'Test_Accuracy']
gap = train_acc - test_acc
print(f"   Écart train-test: {gap:.4f}")
if gap > 0.05:
    print("   ⚠️  Risque de surapprentissage")
else:
    print("   ✅ Bonne généralisation")
```

```python
## Interprétabilité
# Importance XGBoost
if hasattr(xgb_model, 'feature_importances_'):
    # Il est nécessaire d'entraîner xgb_model pour avoir feature_importances_
    xgb_model.fit(X_train_scaled, y_train)
    feature_imp_xgb = pd.DataFrame({
    'feature': X.columns,
    'importance': xgb_model.feature_importances_
}).sort_values('importance')
    plt.figure(figsize=(10, 6))
    plt.barh(feature_imp_xgb['feature'], feature_imp_xgb['importance'])
    plt.title('Importance des variables - XGBoost')
    plt.tight_layout()
    plt.show()
    print('\nTop 5 variables (XGBoost):')
    display(feature_imp_xgb.tail(5).iloc[::-1])
else:
    print('XGBoost n\'a pas d\'attribut feature_importances_')

# SVM linéaire (pour comparaison d'importance uniquement)
svm_linear = SVC(kernel='linear', random_state=123)
svm_linear.fit(X_train_scaled, y_train)
if hasattr(svm_linear, 'coef_'):
    feature_imp_svm = pd.DataFrame({
        'feature': X.columns,
        'importance': np.abs(svm_linear.coef_).mean(axis=0)
    }).sort_values('importance')
    plt.figure(figsize=(10, 6))
    plt.barh(feature_imp_svm['feature'], feature_imp_svm['importance'])
    plt.title('Importance des variables - SVM Linéaire (valeurs absolues coef)')
    plt.tight_layout()
    plt.show()
    print('\nTop 5 variables (SVM linéaire):')
    display(feature_imp_svm.tail(5).iloc[::-1])
```

```python
models = [('XGBoost', xgb_model), ('SVM (RBF)', svm_model)]
for name, model in models:
    train_sizes, train_scores, test_scores = learning_curve(
        model, X_train_scaled, y_train, cv=5, scoring='accuracy', n_jobs=-1, train_sizes=np.linspace(0.1, 1.0, 8)
    )
    plt.figure(figsize=(8, 4))
    plt.plot(train_sizes, np.mean(train_scores, axis=1), 'o-', label='Train')
    plt.plot(train_sizes, np.mean(test_scores, axis=1), 'o-', label='Test')
    plt.title(f"Courbe d'apprentissage - {name}")
    plt.xlabel('Taille du jeu d\'entraînement')
    plt.ylabel('Accuracy')
    plt.legend()
    plt.grid(True, alpha=0.3)
    plt.show()
```

```python
# Création d'un pipeline pour SVM
pipe_svm = Pipeline([
    ('scaler', StandardScaler()),
    ('clf', SVC(random_state=123))
])
```

```python
# Définition de la grille de paramètres
param_grid = {
    'clf__C': np.logspace(-3, 3, 20),  # 20 valeurs de 0.001 à 1000
    'clf__gamma': np.logspace(-4, 1, 20),  # 20 valeurs de 0.0001 à 10
    'clf__kernel': ['rbf', 'linear', 'poly'],  # Test de différents kernels
    'clf__degree': [2, 3, 4],  # Si kernel='poly'
    'clf__class_weight': [None, 'balanced']
}
```

```python
# Configuration du RandomizedSearchCV optimisé
cv = StratifiedKFold(n_splits=5, shuffle=True, random_state=123)

random_search = RandomizedSearchCV(
    pipe_svm,
    param_distributions=param_grid,
    n_iter=20,
    scoring='f1_weighted',
    cv=cv,
    n_jobs=-1,
    verbose=1,
    random_state=123,
    return_train_score=True
)
```

```python
random_search.fit(X_train, y_train)

print('Optimisation terminée!')
print('Meilleurs paramètres trouvés:')
for param, value in random_search.best_params_.items():
    print(f"  {param}: {value}")
print(f"Meilleur score CV (f1_weighted) : {random_search.best_score_:.4f}")

# Récupération du meilleur modèle
best_svm = random_search.best_estimator_

# Évaluation sur le jeu de test
y_pred = best_svm.predict(X_test)

print("\n📊 Rapport de classification (jeu de test) - Modèle Optimisé:")
print(classification_report(y_test, y_pred, zero_division=0))
```

```python
# Calcul du F1-score pour comparaison
f1_optimized = f1_score(y_test, y_pred, average='weighted')
print(f"F1-Score optimisé: {f1_optimized:.4f}")
```

```python
## Matrice de confusion
cm = confusion_matrix(y_test, y_pred)
plt.figure(figsize=(7,5))
sns.heatmap(cm, annot=True, fmt='d', cmap='Blues',
            xticklabels=np.unique(y_test),
            yticklabels=np.unique(y_test))
plt.title(f"Matrice de confusion — SVM optimisé\nF1-test : {f1_optimized:.3f}")
plt.xlabel("Prédictions")
plt.ylabel("Valeurs réelles")
plt.tight_layout()
plt.show()
```

```python
## OPTIMISATION XGBOOST ÉGALEMENT

# Grille pour XGBoost
param_grid_xgb = {
    'n_estimators': [100, 200, 400],
    'max_depth': [3, 5, 7],
    'learning_rate': [0.01, 0.05, 0.1],
    'subsample': [0.7, 0.9, 1.0],
    'colsample_bytree': [0.7, 0.9, 1.0],
    'min_child_weight': [1, 3, 5]
}

xgb_random_search = RandomizedSearchCV(
    XGBClassifier(random_state=123, use_label_encoder=False, eval_metric='mlogloss', n_jobs=-1),
    param_distributions=param_grid_xgb,
    n_iter=40,
    scoring='f1_weighted',
    cv=cv,
    n_jobs=-1,
    verbose=1,
    random_state=123,
    return_train_score=True
)

xgb_random_search.fit(X_train_scaled, y_train)
best_xgb = xgb_random_search.best_estimator_

print('XGBoost optimisé!')
print(f"Meilleur score XGBoost CV: {xgb_random_search.best_score_:.4f}")

```

```python
# Comparaison finale des modèles
final_models = {
    'SVM Optimisé': best_svm,
    'XGBoost Optimisé': best_xgb
}
```

```python
final_results = []
for name, model in final_models.items():
    if name == 'SVM Optimisé':
        # Le SVM utilise déjà le scaler intégré dans le pipeline
        y_pred = model.predict(X_test)
        y_train_pred = model.predict(X_train)
    else:
        # XGBoost utilise les données déjà standardisées
        y_pred = model.predict(X_test_scaled)
        y_train_pred = model.predict(X_train)

    final_results.append({
        'Model': name,
        'Train_Accuracy': accuracy_score(y_train, y_train_pred),
        'Test_Accuracy': accuracy_score(y_test, y_pred),
        'F1-Score': f1_score(y_test, y_pred, average='weighted', zero_division=0),
        'Precision': precision_score(y_test, y_pred, average='weighted', zero_division=0),
        'Recall': recall_score(y_test, y_pred, average='weighted', zero_division=0)
    })

final_df = pd.DataFrame(final_results)
display(final_df)
```

```python
# Trouver le meilleur modèle selon F1-score
best_model_name = final_df.loc[final_df['F1-Score'].idxmax(), 'Model']
best_model = final_models[best_model_name]

# Trouver la meilleure accuracy (ou autre métrique)
best_accuracy = final_df.loc[final_df['F1-Score'].idxmax(), 'Test_Accuracy']

# Trouver les paramètres associés
if best_model_name == 'SVM Optimisé':
    best_params = random_search.best_params_
else:
    best_params = xgb_random_search.best_params_
```

```python
# 11. SAUVEGARDE
model_data = {
    'model': best_model,
    'scaler': scaler,
    'feature_names': X.columns.tolist(),
    'accuracy': best_accuracy,
    'best_params': best_params
}

joblib.dump(model_data, 'juice_model.pkl')
```
